# Loading a Corpus

```python
import nltk
from nltk.book import *
```

```python
#open file for reading and convert text into list of words
f = open('tweets1.txt', 'r')
text = f.read()
text1 = text.split()
abstract = nltk.Text(text1)
f.close()
```

```python
type(abstract)
print(dir(abstract))
```

```python
# concordance on word 'luck'
abstract.concordance('luck')
```
