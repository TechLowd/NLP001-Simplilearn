# Conditional Frequency Distribution
***

Conditional Frequency Distribution is a collection of frequency distributions for same experiment run under different conditions.

##### Getting started

```python
import nltk
from nltk import ConditionalFreqDist
```

```python
l1 = [('cat1', 'good'), ('cat2', 'bad'), ('cat1', 'good')]
```

```python
cfd = ConditionalFreqDist(l1)
```

```python
print(dir(cfd))
```

```python
cfd.items()
```

```python
cfd['cat1']
```

##### Working with NLTK corpus

```python
from nltk.corpus import brown
```

```python
brown.categories()
```

```python
cfd = ConditionalFreqDist((genre, word)
                         for genre in brown.categories()
                         for word in brown.words(categories=genre))

genres = ['adventure', 'editorial', 'government', 'humor', 'lore', 'mystery', 'religion', 'romance']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```

```python
cfd.items()
```

##### Using list comprehension create a list of pair tuples containing sex categories and last letter of name

```python
l_names = ([('male', name[-1]) for name in names.words('male.txt')] +
          [('female', name[-1]) for name in names.words('female.txt')])

print(l_names)
```

```python
cfd = ConditionalFreqDist(l_names)
```

```python
cfd.keys()
```

```python
cfd.items()
```

```python
print(dir(cfd))
```

```python
cfd.plot()
```

- A person is more likely to be female if name ends with letter 'a' or 'e'.
- A person is more likely to be male if name ends with letter 'r', 's', or 't'.
